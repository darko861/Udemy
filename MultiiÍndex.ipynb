## Change Index Level Name with the set_names Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.index.set_names(names = ["Day", "Location"], inplace = True)
```

```python
bigmac.head(3)
```

```python
bigmac.index.set_names(names = "Date", level = 0)
bigmac.index.set_names(names = "Date", level = "Day", inplace = True)
```

```python
bigmac.head(3)
```

## The transpose Method on a MultiIndex DataFrame

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.transpose()
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01", "Sri Lanka") : ("2010-01-01", "Ukraine" )]
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01", ("Sri Lanka" , "Ukraine", "Sweden"))]
```

## The .swaplevel() Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.swaplevel("Date","Country")

bigmac.swaplevel("Country", "Date")
```

## The .pivot() Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
len(sales)
```

```python
sales["Salesman"].value_counts()
```

```python
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

## The .pivot_table() Method

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
pd.pivot_table(data = foods, values = "Spend", index= ["Gender", "Item"], columns = ["City"] , aggfunc = "max")
```
