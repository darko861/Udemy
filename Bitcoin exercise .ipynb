```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from exchanges.coindesk import CoinDesk
```

```python
bitcoin_prices = CoinDesk.get_historical_data_as_dict(start='2013-09-01', end = None)
```

```python
#Create dataframe

df=pd.DataFrame(bitcoin_prices, index=[0])
df
```

```python
#transpose dataframe

df=df.T
```

```python
df
```

```python
#create columns with dates from index
df["Date"]=df.index
df
```

```python
#rename column 0 and change it to price
df.rename(columns={0:'Price'}, inplace=True)
df
```

```python
#Change order of columns -->date, price

df = df[['Date', 'Price']]
df
```

```python
#reset index
df.reset_index(inplace=True)

df
```

```python
df.drop(columns='index', inplace=True)
df
```

```python
plt.plot(df.Price) #call the dataframe df and then the column name
plt.show()
```
